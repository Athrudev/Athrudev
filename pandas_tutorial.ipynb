```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(np.random.rand(300,8),index=np.arange(300))
print(df) #DataFrmae is multi-dimensional
print(type(df),"\n\n")
print(df.describe(),"\n\n")
print(df.T)
```

```python
print(df.head())
```

```python
newdf1=df
print(newdf1) #it will not create a copy of the Dataframe, it will just create an view of the DataFrame(like an pointer)
newdf2=df.copy() #to create an copy of the existing DataFrame we use 'copy()'
print(newdf2)
```
